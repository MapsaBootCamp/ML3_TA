## c. In `walmart_sales`: calculate the average monthly sales by Department for each Store. Which Store(s) has the highest average monthly sales on the department(s)? Report the Store, Department, Date.

````python
store_monthly = (
    sales_iso
    .groupby(['Dept'])
    .apply(
        lambda df: 
        df
        .groupby(['Store'])
        .resample('M')
        .Weekly_Sales
        .mean()
    )
    .to_frame()
).reset_index()
````
