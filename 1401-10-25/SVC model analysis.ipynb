```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_classification
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.svm import SVC
```

```python
X, y = make_classification(
    n_features=2,
    n_redundant=0,
    n_clusters_per_class=2,  # The number of clusters per class.
    flip_y=0.00,  # The fraction of samples whose class is assigned randomly.
    class_sep=1.2,  # Larger values spread out the clusters/classes and make the classification task easier.
    weights=[0.2],
    hypercube=True,
    random_state=42,
    shift=0,
    scale=0.5
)
print(np.bincount(y))
ax = plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
def svc_test(**model_params):
    model = SVC(
        random_state=132,
        **model_params
    ).fit(X, y)
    disp = DecisionBoundaryDisplay.from_estimator(
        model, X, response_method="predict",
        alpha=0.5,
    )
    disp.ax_.scatter(X[:, 0], X[:, 1], c=y, edgecolor="k")

    plt.show()
```

```python
svc_test(
    C=1.0,
    kernel='rbf',
    degree=3,
    gamma='scale',
    shrinking=False,
    class_weight='balanced',
)
```

```python
svc_test(
    C=1.0,
    kernel='rbf',
    degree=3,
    gamma='scale',
    shrinking=False,
    # class_weight='balanced',
)
```

```python
svc_test(
    C=1,
    kernel='poly',
    degree=3,
    gamma='scale',
    shrinking=True,
    # class_weight='balanced',
)
```

```python
svc_test(
    C=1,
    kernel='poly',
    degree=3,
    gamma='scale',
    shrinking=True,
    class_weight='balanced',
)
```
